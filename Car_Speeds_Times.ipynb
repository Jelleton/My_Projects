```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# Here we shall establish our files:
# CR = Car Routes text file, TCD = Traffic Camera Detection text file.
CR = pd.read_csv("CarRoutes.txt", index_col = "Car#")

# Assigning variables based on car numbers and removing unneeded car variables:
CRone = CR.drop(["Car2", "Car3"]).reset_index(drop = True)
CRtwo = CR.drop(["Car1", "Car3"]).reset_index(drop = True)
CRthree = CR.drop(["Car1", "Car2"]).reset_index(drop = True)

# .reset_index(drop = True) resets to default indexing format and removes current index:
```

```python
# Creating a function to plot and get data we need for each car:
def cardata(CR):
    #Time vs Speed:
    x1 = CR[['Time']]
    y1 = CR[['Speed (m/s)']]

    # Latitude vs Longitude:
    x2 = CR[['Longitude']]
    y2 = CR[['Latitude']]

    # Graphing:
    plt.figure(1)
    plt.xlabel('Time (s)')
    plt.ylabel('Speed (m/s)')
    plt.title('Speed vs Time')
    plt.plot(x1,y1,'r-', label = 'Car One')
    plt.legend()
    plt.grid(b=True, which='major', linestyle='--')

    plt.figure(2)
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Longitude vs Latitude')
    plt.plot(x2,y2, label = "Car One")
    plt.legend()
    plt.grid(b=True, which='major', linestyle='--')

    # Max speed of car one:
    MaxSpeed = CR[["Speed (m/s)"]].max().at["Speed (m/s)"]

    # Locating specific row at which car one reaches top speed:
    MaxSpeedIndex = CR[["Speed (m/s)"]].idxmax().at["Speed (m/s)"]
    MaxSpeedTime = CR.loc[MaxSpeedIndex].at["Time"]

    # Printing out the final results:
    print("This was the top speed for car one: " + str(MaxSpeed) + " m/s")
    print("This is when car one reached top speed: " + str(MaxSpeedTime) + " seconds")

cardata(CRone)
# The operator .split() turns parses the strings into a list and .strip() removes string. 
```

```python
cardata(CRtwo)
```

```python
cardata(CRthree)
```
