```python
# Reading out file through pandas so we can manipulate it:
TC = pd.read_csv("TrafficCameraDetections.txt", index_col = "Camera")

# Separating the Cameras into their own variables:
TCone = TCD.drop(["TrafficCamera_2", "TrafficCamera_3", "TrafficCamera_4", "TrafficCamera_5"]).reset_index(drop = True)
TCtwo = TCD.drop(["TrafficCamera_1", "TrafficCamera_3", "TrafficCamera_4", "TrafficCamera_5"]).reset_index(drop = True)
TCthree = TCD.drop(["TrafficCamera_2", "TrafficCamera_1", "TrafficCamera_4", "TrafficCamera_5"]).reset_index(drop = True)
TCfour = TCD.drop(["TrafficCamera_2", "TrafficCamera_3", "TrafficCamera_1", "TrafficCamera_5"]).reset_index(drop = True)
TCfive = TCD.drop(["TrafficCamera_2", "TrafficCamera_3", "TrafficCamera_4", "TrafficCamera_1"]).reset_index(drop = True)
```

```python
# Function for finding whether or not camera detected a certain car:
def detection(Camera):
    cars = ["Car1", "Car2", "Car3"]
    value = Camera['Car#'].value_counts()
    for i in cars:
        if i in value:
            print(str(i) + " was detected by camera.")
        else:
            print(str(i) + " was not detected by camera.")
    print("This is how long (seconds) the camera tracked each car: \n" + str(value).strip("Name:, dtype int64"))
```

```python
# Checking what cars each camera detected:
Cameras = {"Camera One":TCone, "Camera Two":TCtwo, "Camera Three":TCthree, "Camera Four":TCfour, "Camera Five":TCfive}

# Iterating the function to compact code:
for key, value in Cameras.items():
    print(key + " detected these cars: ")
    detection(value)
    print("\n")
```
